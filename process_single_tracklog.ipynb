```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/home/ganesh/geo/tracklog-A1F70AB6-D3F1-4FF3-AF7E-EB0EF3CC21AB.csv', skiprows=2)
df
```

```python
plt.figure()
plt.plot(df['Speed'])
plt.xlabel('Elapsed Time (s)')
plt.ylabel('Groundspeed (kt)')
plt.title('Groundspeed over Time')
```

```python
fig = plt.figure()
ax = fig.add_subplot()
scatter = ax.scatter(df['Longitude'], df['Latitude'], c=df['Altitude'], cmap='viridis', s=5)
fig.colorbar(scatter, shrink=0.5, aspect=5, label='Altitude')
ax.set_title('Flight Path')
```
